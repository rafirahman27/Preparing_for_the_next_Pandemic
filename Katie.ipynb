## Creating a function

```python
cancer = pd.read_csv('Ignore/Cancer.csv')
```

```python
heart = pd.read_csv('Ignore/Cardiovascular.csv')
```

```python
copd = pd.read_csv('Ignore/COPD.csv')
```

```python
def cleaning(df):
    print(df.shape)
    df = df[df['YearStart'] >= 2019]
    print(df.shape)
    df = df[['YearStart', 'LocationAbbr', 'LocationDesc', 'Question', 'DataValue', 'Stratification1']]
    print(df.shape)
    df.drop_duplicates(inplace=True)
    return df  
               
```

```python
cleaning(cancer)
```

```python
cleaning(heart)
```

```python
cleaning(copd)
```

```python
diabetes = pd.read_csv('Ignore/Diabetes.csv')
```

```python
cleaning(diabetes)
```
